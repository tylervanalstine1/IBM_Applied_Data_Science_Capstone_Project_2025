```python
# Task 2: Mark the success/failed launches for each site on the map


# Create the map centered at the average launch location
site_map = folium.Map(location=[spacex_df['Lat'].mean(), spacex_df['Long'].mean()], zoom_start=5)

# Add marker cluster to group nearby launch points
marker_cluster = MarkerCluster().add_to(site_map)

# Add success/failure markers
for i, row in spacex_df.iterrows():
    location = [row['Lat'], row['Long']]
    result = 'Success' if row['class'] == 1 else 'Failure'
    color = 'green' if row['class'] == 1 else 'red'

    folium.Marker(
        location=location,
        popup=f"{row['Launch Site']} - {result}",
        icon=folium.Icon(color=color, icon='ok-sign')
    ).add_to(marker_cluster)

# Add mouse position display
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
site_map.add_child(MousePosition(
    position='topright',
    separator=' | ',
    prefix='Lat | Long:',
    lat_formatter=formatter,
    lng_formatter=formatter
))

# Show the map (inline if possible)
site_map
```

```python
import pandas as pd
import math

# Haversine function 
def haversine(coord1, coord2):
    R = 6371  # Earth radius in kilometers
    lat1, lon1 = coord1
    lat2, lon2 = coord2
    phi1 = math.radians(lat1)
    phi2 = math.radians(lat2)
    dphi = math.radians(lat2 - lat1)
    dlambda = math.radians(lon2 - lon1)
    a = math.sin(dphi/2)**2 + math.cos(phi1) * math.cos(phi2) * math.sin(dlambda/2)**2
    return R * (2 * math.atan2(math.sqrt(a), math.sqrt(1 - a)))

# Define proximity points
proximities = {
    'Coastline': (28.5633, -80.5678),
    'City': (29.7604, -95.3698),     # Houston, TX
    'Railway': (28.5721, -80.5852),
    'Highway': (28.5619, -80.5770)
}

# Calculate distances using haversine
distance_data = []
for _, site in launch_sites_df.iterrows():
    site_coord = (site['Lat'], site['Long'])
    row = {'Launch Site': site['Launch Site']}
    for name, prox_coord in proximities.items():
        row[f'Distance to {name} (km)'] = round(haversine(site_coord, prox_coord), 2)
    distance_data.append(row)

# Create and display distance DataFrame
distances_df = pd.DataFrame(distance_data)
distances_df
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Coordinates of launch site and closest coastline point (from MousePosition)
launch_site_lat = 28.563197
launch_site_lon = -80.576820
coastline_lat = 28.5637
coastline_lon = -80.5650

distance_coastline = calculate_distance(28.563197, -80.576820, 28.5637, -80.567)
print(f"Distance to coastline: {distance_coastline:.2f} km")

```
